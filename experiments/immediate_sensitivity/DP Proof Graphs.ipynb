```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns

import numpy as np
sns.set(style="white", palette="muted", color_codes=True, context="talk")
```

# SHALLOW network

```python
ours_epsilons = [0.06785389680889561, 0.1578538968088956, 1.0578538968088955, 10.057853896808895, 100.0578538968089]
ours_means = [0.767, 0.8540000000000001, 0.9109999999999999, 0.9195, 0.922]
ours_stds = [0.05933801479658718, 0.09481033698917013, 0.014456832294800973, 0.014396180048887966, 0.03083828789021856]
```

```python
baseline_epsilons = [0.06785389680889561, 0.1578538968088956, 1.0578538968088955, 10.057853896808895, 100.0578538968089]
baseline_means = [0.691, 0.7070000000000001, 0.8485000000000001, 0.866, 0.922]
baseline_stds = [0.15145626431415768, 0.17461672313956644, 0.06261189982742899, 0.13421624342828256, 0.00979795897113272]
```

```python
plt.figure(figsize=(7, 5), dpi=80)

markers, caps, bars = plt.errorbar(baseline_epsilons, baseline_means, 
                                   yerr=baseline_stds, 
                                   label='Gradient Clipping')
[bar.set_alpha(0.5) for bar in bars]
[cap.set_alpha(0.5) for cap in caps]

markers, caps, bars = plt.errorbar(ours_epsilons, ours_means, 
                                   yerr=ours_stds, 
                                   label='Our Approach')
[bar.set_alpha(0.5) for bar in bars]
[cap.set_alpha(0.5) for cap in caps]

plt.xlabel('Epsilon')
plt.ylabel('Accuracy')
plt.title('Synthetic Data')
plt.axhline(y=0.945, color='k', linestyle='--', label='Non-private baseline')
plt.ylim(0.4, 1.0)
# plt.xlim(0.0, 0.3)
plt.xscale('log')
plt.xticks()
plt.legend();
plt.tight_layout()
plt.savefig('initial_results_proof_shallow.png')
```
